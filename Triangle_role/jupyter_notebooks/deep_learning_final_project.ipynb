```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
#  Import and read the csv.
data_df = pd.read_csv("Resources/sample_data.csv")
data_df.head()
```

```python
#Using drop() to delete rows based on column value
data_df.drop(data_df[data_df['Year'] < 2019].index, inplace = True)
data_df.head()
```

```python
# removing country, state due to FIPS being good representation
# getting 0.0 accuracy. removing WA, BA, IA, AA, TOM, NH, H also
data_df = data_df.drop(columns=["C_S", "WA","BA","IA","AA","NA","TOM","NH","H"])
data_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = data_df["Cost"].values
X = data_df.drop(["Cost"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 4
hidden_nodes_layer2 = 2

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
# X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# Compile the model
# trying mean_squared_error and mse instead of binary_crossentropy and accuracy, respectively
# i realized these are the same measurement tools
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
