```python
import pandas as pd
C2020 = pd.read_csv('COVID2020.csv')
C2021 = pd.read_csv('COVID2021.csv')
```

# COVID data

## 2020

```python
C2020
```

```python
C2020['date']=pd.to_datetime(C2020['date'],format="%m/%d/%Y")
```

```python
C2020['fips'] = C2020['fips'].fillna(0)
```

```python
C2020 = C2020[(C2020['fips'] <= 56045) & (C2020['fips']>0)]
C2020.sort_values(by=["fips"])
C2020
```

```python
C2020['fips']=C2020['fips'].astype(str)
```

```python
C2020['fips']=C2020['fips'].str.split('.').str[0]
C2020['fips']=C2020['fips'].str.pad(5,'left','0')
C2020
```

```python
C2020.sort_values(by=["fips"])
```

```python
COVID2020=C2020.groupby(by=['fips',(C2020.date.dt.month)])[["cases"]].sum()
COVID2020
```

```python
COVID2020.to_csv("cleancovid2020.csv")
```

# Income Data

```python
income2017=pd.read_csv('ACSDT5Y2017.B19301-Data.csv')
income2018=pd.read_csv('ACSDT5Y2018.B19301-Data.csv')
income2019=pd.read_csv('ACSDT5Y2019.B19301-Data.csv')
income2020=pd.read_csv('ACSDT5Y2020.B19301-Data.csv')
income2021=pd.read_csv('ACSDT5Y2021.B19301-Data.csv')
```

```python
income2017['fips']=income2017['GEO_ID'].str.split('US').str[-1]
income2018['fips']=income2018['GEO_ID'].str.split('US').str[-1]
income2019['fips']=income2019['GEO_ID'].str.split('US').str[-1]
income2020['fips']=income2020['GEO_ID'].str.split('US').str[-1]
income2021['fips']=income2021['GEO_ID'].str.split('US').str[-1]
```

```python
#rename percapita column, get rid of top row and only keep nesseary columns
income2017.rename(columns={income2017.columns[2]:"PCIncome"},inplace=True)
income2017=income2017[["fips","PCIncome"]]
income2017 = income2017.iloc[1: , :]

income2018.rename(columns={income2018.columns[2]:"PCIncome"},inplace=True)
income2018=income2018[["fips","PCIncome"]]
income2018 = income2018.iloc[1: , :]

income2019.rename(columns={income2019.columns[2]:"PCIncome"},inplace=True)
income2019=income2019[["fips","PCIncome"]]
income2019 = income2019.iloc[1: , :]

income2020.rename(columns={income2020.columns[2]:"PCIncome"},inplace=True)
income2020=income2020[["fips","PCIncome"]]
income2020 = income2020.iloc[1: , :]

income2021.rename(columns={income2021.columns[2]:"PCIncome"},inplace=True)
income2021=income2021[["fips","PCIncome"]]
income2021 = income2021.iloc[1: , :]
```

```python
income2017['fips']=income2017['fips'].str.pad(5,'left','0')
income2018['fips']=income2018['fips'].str.pad(5,'left','0')
income2019['fips']=income2019['fips'].str.pad(5,'left','0')
income2020['fips']=income2020['fips'].str.pad(5,'left','0')
income2021['fips']=income2021['fips'].str.pad(5,'left','0')
```

```python
income2017["year"]="2017"
income2018["year"]="2018"
income2019["year"]="2019"
income2020["year"]="2020"
income2021["year"]="2021"
```

```python
income2017
```

```python
income2017.to_csv("income2017.csv")
income2018.to_csv("income2018.csv")
income2019.to_csv("income2019.csv")
income2020.to_csv("income2020.csv")
income2021.to_csv("income2021.csv")
```

```python
income=pd.concat(
    [income2017,income2018,income2019,income2020,income2021],
    axis=0,
    join="outer",
    ignore_index=False,
    keys=["2017",'2018','2019','2020','2021'],
    levels=None,
    names=None,
    verify_integrity=False,
    copy=True,
)
```

```python
income
```

```python
income.to_csv("incomeall.csv")
```

# Race/Eth

```python
race=pd.read_csv('CC-EST2020-ALLDATA6.csv',encoding = "ISO-8859-1")
race.head()
```

```python
race.dtypes
```

```python
race["YEAR"]=race['YEAR'].replace(10,2017)
race["YEAR"]=race['YEAR'].replace(11,2018)
race["YEAR"]=race['YEAR'].replace(12,2019)
```

```python
race=race[race.YEAR >=2017]
race=race[race.YEAR <=2019]
race.head()
```

```python
race['C_S']=race['CTYNAME'].astype(str)+','+race['STNAME'].astype(str)
race.head()
```

```python
race['WA']=race['WA_MALE'].astype(int)+race['WA_FEMALE'].astype(int)
race['BA']=race['BA_MALE'].astype(int)+race['BA_FEMALE'].astype(int)
race['IA']=race['IA_MALE'].astype(int)+race['IA_FEMALE'].astype(int)
race['AA']=race['AA_MALE'].astype(int)+race['AA_FEMALE'].astype(int)
race['NA']=race['NA_MALE'].astype(int)+race['NA_FEMALE'].astype(int)
race['TOM']=race['TOM_MALE'].astype(int)+race['TOM_FEMALE'].astype(int)
race['NH']=race['NH_MALE'].astype(int)+race['NH_FEMALE'].astype(int)
race['H']=race['H_MALE'].astype(int)+race['H_FEMALE'].astype(int)
```

```python
raceeth1=race[['YEAR','C_S','WA','BA','IA','AA','NA','TOM','NH','H']]
raceeth1
```

```python
race=pd.read_csv('cc-est2021-all.csv',encoding = "ISO-8859-1")
race.head()
```

```python
race["YEAR"]=race['YEAR'].replace(2,2020)
race["YEAR"]=race['YEAR'].replace(3,2021)
```

```python
race=race[race.YEAR >=2017]
race=race[race.AGEGRP==0]
race.head()
```

```python
race['C_S']=race['CTYNAME'].astype(str)+','+race['STNAME'].astype(str)
race['WA']=race['WA_MALE'].astype(int)+race['WA_FEMALE'].astype(int)
race['BA']=race['BA_MALE'].astype(int)+race['BA_FEMALE'].astype(int)
race['IA']=race['IA_MALE'].astype(int)+race['IA_FEMALE'].astype(int)
race['AA']=race['AA_MALE'].astype(int)+race['AA_FEMALE'].astype(int)
race['NA']=race['NA_MALE'].astype(int)+race['NA_FEMALE'].astype(int)
race['TOM']=race['TOM_MALE'].astype(int)+race['TOM_FEMALE'].astype(int)
race['NH']=race['NH_MALE'].astype(int)+race['NH_FEMALE'].astype(int)
race['H']=race['H_MALE'].astype(int)+race['H_FEMALE'].astype(int)
```

```python
raceeth2=race[['YEAR','C_S','WA','BA','IA','AA','NA','TOM','NH','H']]
raceeth2
```

```python
raceeth=pd.concat(
    [raceeth1,raceeth2],
    axis=0,
    join="outer",
    ignore_index=False,
    keys=None,
    levels=None,
    names=None,
    verify_integrity=False,
    copy=True,
)
```

```python
raceeth.to_csv("RaceEthnicity.csv")
```
